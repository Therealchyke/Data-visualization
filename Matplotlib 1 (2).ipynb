```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.set_title('Title')
plt.plot(x,y)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
plt.plot(x,y)
ax2=fig.add_axes( [0.2,0.5,.2,.2])
plt.plot(x,y)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax1.set_ylabel('Z')
ax1.set_xlabel('X')
ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])
plt.plot(x,z)

ax2=fig.add_axes([0.2,0.5,.4,.4])
ax2.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
plt.plot(x,y)

```

```python
fig, axes=plt.subplots(nrows=1, ncols=2)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_ylim([0,200])
axes[0].plot(x,y,color='b',lw=2.5,ls='--')

axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
axes[1].set_ylim([0,10000])
axes[1].plot(x,z,color='r',lw=3)
plt.tight_layout()
```

```python
fig, axes=plt.subplots(nrows=1, ncols=2, figsize=(12,2))
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_ylim([0,200])
axes[0].plot(x,y,color='b',lw=2.5,ls='-')

axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
axes[1].set_ylim([0,10000])
axes[1].plot(x,z,color='r',lw=3, ls='--')
plt.tight_layout()
```
